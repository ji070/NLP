```python
text = """Dear Amazon, last week I ordered an Optimus Prime action figure \
from your online store in Germany. Unfortunately, when I opened the package, \
I discovered to my horror that I had been sent an action figure of Megatron \
instead! As a lifelong enemy of the Decepticons, I hope you can understand my \
dilemma. To resolve the issue, I demand an exchange of Megatron for the \
Optimus Prime figure I ordered. Enclosed are copies of my records concerning \
this purchase. I expect to hear from you soon. Sincerely, Bumblebee."""
```

### Text Classification

Î¨∏Ïû•ÏùÑ Ïñ¥Îñ§Í±∏Î°ú Í∑úÏ†ïÌïòÎäîÍ±∞Îã§ // ÎÜìÏπ®

```python
#hide_output
from transformers import pipeline

classifier = pipeline("text-classification")  # ÏÉùÏÑ±ÌïòÎäî ÌÖåÏä§ÌÅ¨ Ïù¥Î¶ÑÏù¥Îã§., Î™®Îç∏Ïù¥Î¶ÑÎì±Îì± Ïó¨Îü¨Í∞úÎ•º ÏßÄÏ†ïÌï† Ïàò ÏûàÎã§. # Î≥¥ÌÜµ tokenizerÎèÑ Í∞ôÏù¥ Î∞∞Ìè¨ÌïúÎã§. # vocaÎäî ÏÇ¨Ï†ÑÏóê ÏûàÎäî Îã®Ïñ¥Îã§.
```

```python
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

### Named Entity Recognition

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
# Í∞úÏ≤¥Î™ÖÏúºÎ°ú Ïù∏ÏãùÌïúÍ≤ÉÎßå Ï∂úÎ†•ÌïúÎã§. start,endÎäî ÏúÑÏπòÎ•º ÎÇòÌÉÄÎÇ∏Îã§. Ïù∏Îç±Ïä§Í∏∞Ï§ÄÏúºÎ°ú endÎäî ÏßÅÏ†ÑÍπåÏßÄÎ°ú Ïù∏Ïãù
# ÌÜ†ÌÅ∞ÎßàÎã§ ÏòàÏ∏°ÏùÑ ÌïúÎã§.
# Ïñ¥Îñ§ ÌÜ†ÌÅ∞ÏùÄ Îã®Ïñ¥Í∞Ä ÌïòÎÇò ÏùºÏßÄÎùºÎèÑ ÎëêÍ∞úÏùò ÌÜ†ÌÅ∞ÏúºÎ°ú Î∂ÑÎ¶¨ ÎêòÎäî Í≤ΩÏö∞ÎèÑ ÏûàÎã§. 3Í≥º 7ÏùÑ Î≥¥Î©¥ , 3Í≥º4Î•º Î≥¥Î©¥ Îã§Î•¥Í≤å Î∂ÑÎ¶¨ ÎêòÏñ¥ÏûàÎã§. MegatronÏùÑ  Mega, ##tronÏúºÎ°ú Î∂ÑÎ¶¨ ÌñàÍ≥† ÏõåÎìú Í∏∞Ï§ÄÏúºÎ°úÎäî ÎèôÏùºÌïúÎç∞ ÌÜ†ÌÅ∞ Í∏∞Ï§ÄÏúºÎ°úÎäî ÎÇòÎâòÍ≥† ÎëòÎã§ MISCÎ°ú Î∂ÑÎ¶¨ ÌñàÏúºÎ©¥ ÌïòÎÇòÎ°ú Î≥¥Ïó¨ÏßÄÍ≥† ÌÜ†ÌÇ®Î≥ÑÎ°ú Îã§Î•¥Í≤å Î∂ÑÎ¶¨ ÌñàÏúºÎ©¥ ÌÜ†ÌÅ∞ÎßàÎã§ Î≥¥Ïó¨Ï§ÄÎã§.
```

### Question Answering
ÏàòÏóÖÏóêÏÑú Îã§Î£®ÏßÄ ÏïäÏùÑÍ±∞Îã§ Ïò§ÎûòÎêòÏñ¥ÏÑú

```python
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context=text) # contextÏóêÏÑú ÏßàÎ¨∏Ïóê ÎåÄÌïú ÎãµÏù¥ÎùºÍ≥† ÏòàÏÉÅÎêòÎäî Í≤ÉÏùÑ Ï∞æÏïÑÏÑú Í∑∏Î∂ÄÎ∂ÑÎßå Ïù∏Îç±Ïä§Î°ú Í∞ÄÏ†∏Ïò®Îã§.# contextÏóêÏÑú ÌÜ†ÌÅ∞Ïù¥ startÏùºÌôïÎ•†, endÏùº ÌôúÎ•† ÏòàÏ∏°ÌïòÎã§ // Ïù¥Í≤ÉÎèÑ Ïù¥Í≤å ÎßûÎÇò? ÎÜìÏπ®
pd.DataFrame([outputs])
```

### Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=45, clean_up_tokenization_spaces=True) # ÏµúÎåÄÍ∏∏Ïù¥Î•º 45ÌÜ†ÌÅ∞ Í∏∏Ïù¥Î°ú ÌñàÏúºÎ©¥ Ï¢ãÍ≤†Îã§. Ï§ëÍ∞ÑÍ≤∞Í≥ºÎ•º Ï†ïÎ¶¨ÌïòÏûê(clean_...)
print(outputs[0]['summary_text'])
```

### Translation

```python
translator = pipeline("translation_en_to_de",
                      model="Helsinki-NLP/opus-mt-en-de") # translation_en_to_de: ÏòÅÏñ¥Î•º ÎèÖÏùºÏñ¥Î°ú Î∞îÍæ∏ÎäîÍ±∞ÎùºÍ≥† translationÎí§Ïóê Ï†ëÎØ∏ÏÇ¨Î•º Î∂ôÏòÄÎã§.
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text']) # ÎèÖÏùºÏñ¥Î°ú Î≤àÏó≠ÌïúÍ±∞Îã§.
```

### Text Generation

```python
#hide
from transformers import set_seed
set_seed(42) # Set the seed to get reproducible results # random seedÍ∞Ä Ï§ëÏöîÌïòÎã§
```

```python
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text + "\n\nCustomer service response:\n" + response # ÏùëÎãµÏúºÎ°ú ÌïòÍ≥† Ïã∂ÏùÄ Í∏ÄÏßìÍ∏∞ ÏùºÎ∂ÄÎ•º Ï†ÅÏñ¥ÎÜìÏïòÎã§.
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
